```python
# Data
hours = [0,1,2,3,3,5,5,6,7,7,10,5]
marks = [96,85,82,74,95,68,76,58,65,75,50,84]
```

```python
n = len(hours)
sum_x = sum(hours)
sum_y = sum(marks)
```

```python
sum_x2 = sum(x**2 for x in hours)
```

```python
sum_xy = sum(hours[i] * marks[i] for i in range(n))
```

```python
# Slope (m)
m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
```

```python
# Intercept (c)
c = (sum_y * sum_x2 - sum_x * sum_xy) / (n * sum_x2 - sum_x**2)
```

```python
print(f"Fitted line: y^ = {m:.2f}x + {c:.2f}")
```

```python
# Predict marks for each hour
x=float (input("enter the value of x :"))

```

```python
y_pred = m * (x) + c 
```

```python
print("pridicted value for given x :",y_pred)
```
